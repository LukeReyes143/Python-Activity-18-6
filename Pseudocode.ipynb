```python
# Data structures
books = []  # List of dictionaries to store book details
members = []  # List of dictionaries to store member details
lending_details = []  # List of dictionaries to store lending details

from IPython.display import display, clear_output
from datetime import date

## 1. Add Book
def add_book():
    clear_output()
    isbn = input("Enter ISBN: ")
    title = input("Enter title: ")
    author = input("Enter author: ")
    genre = input("Enter genre: ")
    availability = input("Enter availability status (available/unavailable): ")
    
    book = {
        "isbn": isbn,
        "title": title,
        "author": author,
        "genre": genre,
        "availability": availability
    }
    
    books.append(book)
    print("Book added successfully.")

## 2. Delete Book
def delete_book():
    clear_output()
    isbn = input("Enter ISBN of the book to be deleted: ")
    
    for book in books:
        if book["isbn"] == isbn:
            books.remove(book)
            print("Book deleted successfully.")
            return
    
    print("Book not found.")

## 3. Display Book Details
def display_book_details():
    clear_output()
    if not books:
        print("No books available.")
        return
    
    for book in books:
        status = "Available" if book["availability"] == "available" else "Lent Out"
        print(f"ISBN: {book['isbn']}, Title: {book['title']}, Author: {book['author']}, Genre: {book['genre']}, Status: {status}")

## 4. Register Member
def register_member():
    clear_output()
    member_id = input("Enter member ID: ")
    name = input("Enter name: ")
    contact = input("Enter contact information: ")
    
    member = {
        "member_id": member_id,
        "name": name,
        "contact": contact
    }
    
    members.append(member)
    print("Member registered successfully.")

## 5. Lend Book
def lend_book():
    clear_output()
    isbn = input("Enter ISBN of the book to be lent: ")
    member_id = input("Enter member ID: ")
    
    # Check if the book is available
    for book in books:
        if book["isbn"] == isbn and book["availability"] == "available":
            # Check if the member exists
            for member in members:
                if member["member_id"] == member_id:
                    book["availability"] = "unavailable"
                    lending_detail = {
                        "isbn": isbn,
                        "member_id": member_id,
                        "lending_date": date.today().strftime("%Y-%m-%d")
                    }
                    lending_details.append(lending_detail)
                    print("Book lent successfully.")
                    return
            print("Member not found.")
            return
    
    print("Book not available for lending.")

## 6. Return Book
def return_book():
    clear_output()
    isbn = input("Enter ISBN of the book to be returned: ")
    
    # Check if the book is lent out
    for lending_detail in lending_details:
        if lending_detail["isbn"] == isbn:
            # Mark the book as available
            for book in books:
                if book["isbn"] == isbn:
                    book["availability"] = "available"
            
            # Remove the lending detail
            lending_details.remove(lending_detail)
            print("Book returned successfully.")
            return
    
    print("Book is not currently lent out.")

# Function to display the main menu
def display_menu():
    clear_output()
    print("Library Management System")
    print("1. Add Book")
    print("2. Delete Book")
    print("3. Display Book Details")
    print("4. Register Member")
    print("5. Lend Book")
    print("6. Return Book")
    print("7. Exit")
    choice = input("Enter your choice: ")
    return choice

# Main program loop
while True:
    choice = display_menu()

    if choice == "1":
        add_book()
    elif choice == "2":
        delete_book()
    elif choice == "3":
        display_book_details()
    elif choice == "4":
        register_member()
    elif choice == "5":
        lend_book()
    elif choice == "6":
        return_book()
    elif choice == "7":
        print("Exiting the program...")
        break
    else:
        print("Invalid choice. Please try again.")
```
